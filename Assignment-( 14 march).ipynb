```python
"""Q1. Write a Python function that takes in two arrays of data and calculates the F-value for a variance ratio
test. The function should return the F-value and the corresponding p-value for the test."""

#Ans:-

import numpy as np
import scipy.stats as stats

def calculate_f_value(data1, data2):
   
    data1 = np.array(data1)
    data2 = np.array(data2)

    variance1 = np.var(data1, ddof=1)
    variance2 = np.var(data2, ddof=1)

    f_value = variance1 / variance2

    df1 = len(data1) - 1
    df2 = len(data2) - 1

    p_value = 1 - stats.f.cdf(f_value, df1, df2)

    return f_value, p_value
data1 = [1, 2, 3, 4, 5]
data2 = [6, 7, 8, 9, 10]

f_value, p_value = calculate_f_value(data1, data2)

print("F-value:", f_value)
print("p-value:", p_value)
```

```python
"""Q2. Given a significance level of 0.05 and the degrees of freedom for the numerator and denominator of an
F-distribution, write a Python function that returns the critical F-value for a two-tailed test."""

#Ans:-import scipy.stats as stats

def get_critical_f_value(significance_level, df_numerator, df_denominator):
   
    critical_f_value = stats.f.ppf(1 - significance_level / 2, df_numerator, df_denominator)

    return critical_f_value

significance_level = 0.05
df_numerator = 3
df_denominator = 10

critical_f_value = get_critical_f_value(significance_level, df_numerator, df_denominator)

print("Critical F-value:", critical_f_value)
```

```python
"""Q3. Write a Python program that generates random samples from two normal distributions with known
variances and uses an F-test to determine if the variances are equal. The program should output the Fvalue, degrees of freedom, and p-value for the test."""

#Ans:-

import numpy as np
import scipy.stats as stats

def perform_f_test(sample1, sample2, variance1, variance2):

    f_value = np.var(sample1, ddof=1) / np.var(sample2, ddof=1)

    degrees_of_freedom = (len(sample1) - 1, len(sample2) - 1)

    p_value = 1 - stats.f.cdf(f_value, *degrees_of_freedom)
    p_value *= 2  

    return f_value, degrees_of_freedom, p_value


variance1 = 10
variance2 = 15
sample_size = 50

sample1 = np.random.normal(loc=0, scale=np.sqrt(variance1), size=sample_size)
sample2 = np.random.normal(loc=0, scale=np.sqrt(variance2), size=sample_size)

f_value, degrees_of_freedom, p_value = perform_f_test(sample1, sample2, variance1, variance2)

print("F-value:", f_value)
print("Degrees of Freedom:", degrees_of_freedom)
print("P-value:", p_value)
```

```python
"""Q4.The variances of two populations are known to be 10 and 15. A sample of 12 observations is taken from
each population. Conduct an F-test at the 5% significance level to determine if the variances are
significantly different."""

#ans:-

import scipy.stats as stats

variance1 = 10
variance2 = 15
sample_size1 = 12
sample_size2 = 12

test_statistic = variance2 / variance1

df1 = sample_size1 - 1
df2 = sample_size2 - 1
critical_value = stats.f.ppf(1 - 0.05, df1, df2)

p_value = 1 - stats.f.cdf(test_statistic, df1, df2)
p_value *= 2  

if test_statistic > critical_value:
    print("Reject the null hypothesis. The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis. The variances are not significantly different.")

print("Test Statistic:", test_statistic)
print("Critical Value:", critical_value)
print("P-value:", p_value)
```

```python
"""Q5. A manufacturer claims that the variance of the diameter of a certain product is 0.005. A sample of 25
products is taken, and the sample variance is found to be 0.006. Conduct an F-test at the 1% significance
level to determine if the claim is justified."""

#Ans:-

import scipy.stats as stats

sample_variance = 0.006
claimed_variance = 0.005
sample_size = 25

test_statistic = sample_variance / claimed_variance

df1 = sample_size - 1
df2 = sample_size - 1
critical_value = stats.f.ppf(1 - 0.01/2, df1, df2)

p_value = 1 - stats.f.cdf(test_statistic, df1, df2)
p_value *= 2  

if test_statistic > critical_value:
    print("Reject the null hypothesis. The variance claim is not justified.")
else:
    print("Fail to reject the null hypothesis. The variance claim is justified.")

print("Test Statistic:", test_statistic)
print("Critical Value:", critical_value)
print("P-value:", p_value)


```

```python
"""Q6. Write a Python function that takes in the degrees of freedom for the numerator and denominator of an
F-distribution and calculates the mean and variance of the distribution. The function should return the
mean and variance as a tuple."""

#Ans:-
def calculate_f_distribution_stats(df_numerator, df_denominator):
   
    if df_numerator <= 0 or df_denominator <= 0:
        raise ValueError("Degrees of freedom must be greater than zero.")

    if df_denominator > 2:
        mean = df_denominator / (df_denominator - 2)
        variance = (2 * (df_denominator ** 2) * (df_numerator + df_denominator - 2)) / (
                df_numerator * (df_denominator - 2) ** 2 * (df_denominator - 4))
    else:
        raise ValueError("Degrees of freedom for the denominator must be greater than 2.")

    return mean, variance

```

```python
"""Q7. A random sample of 10 measurements is taken from a normal population with unknown variance. The
sample variance is found to be 25. Another random sample of 15 measurements is taken from another
normal population with unknown variance, and the sample variance is found to be 20. Conduct an F-test
at the 10% significance level to determine if the variances are significantly different."""

#Ans:-

To conduct an F-test at the 10% significance level to determine if the variances of two populations are significantly different, we follow these steps:

Step 1: Define the null hypothesis (H0) and the alternative hypothesis (H1).

H0: The variances of the two populations are equal.
H1: The variances of the two populations are significantly different.

Step 2: Calculate the sample variances and degrees of freedom.

For the first sample:
Sample size (n1) = 10
Sample variance (s1^2) = 25
Degrees of freedom for the first sample (df1) = n1 - 1 = 10 - 1 = 9

For the second sample:
Sample size (n2) = 15
Sample variance (s2^2) = 20
Degrees of freedom for the second sample (df2) = n2 - 1 = 15 - 1 = 14

Step 3: Calculate the F-statistic.

The F-statistic is calculated using the formula:

F = s1^2 / s2^2

In this case, F = 25 / 20 = 1.25

Step 4: Determine the critical value of the F-distribution.

To determine if the variances are significantly different, we need to compare the calculated F-statistic to the critical value of the F-distribution at the given significance level.

Since we are conducting the F-test at the 10% significance level, we need to find the critical F-value with df1 = 9 and df2 = 14.

Using a statistical table or software, we find that the critical F-value at the 10% significance level for df1 = 9 and df2 = 14 is approximately 2.38.

Step 5: Compare the calculated F-statistic with the critical F-value.

If the calculated F-statistic is greater than the critical F-value, we reject the null hypothesis (H0) and conclude that the variances are significantly different. Otherwise, if the calculated F-statistic is less than or equal to the critical F-value, we fail to reject the null hypothesis and conclude that there is not enough evidence to suggest a significant difference in variances.

In this case, the calculated F-statistic is 1.25, and the critical F-value at the 10% significance level is 2.38. Since 1.25 < 2.38, we fail to reject the null hypothesis. Therefore, we conclude that there is not enough evidence to suggest a significant difference in variances between the two populations at the 10% significance level.




```

```python
"""Q9. The following data represent the test scores of two groups of students: Group A: 80, 85, 90, 92, 87, 83;
Group B: 75, 78, 82, 79, 81, 84. Conduct an F-test at the 1% significance level to determine if the variances
are significantly different."""

#Ans:-

To conduct an F-test to determine if the variances of two groups are significantly different, we need to follow these steps:

Step 1: Define the null hypothesis (H0) and the alternative hypothesis (H1).

The null hypothesis states that the variances of the two groups are equal:
H0: σ₁² = σ₂²

The alternative hypothesis states that the variances of the two groups are significantly different:
H1: σ₁² ≠ σ₂²

Step 2: Calculate the sample variances for each group.

For Group A:
n₁ = 6 (number of data points)
x₁ = {80, 85, 90, 92, 87, 83} (data points)

The sample variance for Group A can be calculated using the formula:

s₁² = Σ(x₁ - x̄₁)² / (n₁ - 1)

where x̄₁ is the sample mean of Group A.

x̄₁ = (80 + 85 + 90 + 92 + 87 + 83) / 6 = 87.83

s₁² = [(80 - 87.83)² + (85 - 87.83)² + (90 - 87.83)² + (92 - 87.83)² + (87 - 87.83)² + (83 - 87.83)²] / (6 - 1)
= [45.55 + 7.61 + 7.61 + 16.81 + 0.69 + 22.03] / 5
= 100.06 / 5
= 20.01

For Group B:
n₂ = 6 (number of data points)
x₂ = {75, 78, 82, 79, 81, 84} (data points)

The sample variance for Group B can be calculated using the same formula:

s₂² = Σ(x₂ - x̄₂)² / (n₂ - 1)

where x̄₂ is the sample mean of Group B.

x̄₂ = (75 + 78 + 82 + 79 + 81 + 84) / 6 = 79.83

s₂² = [(75 - 79.83)² + (78 - 79.83)² + (82 - 79.83)² + (79 - 79.83)² + (81 - 79.83)² + (84 - 79.83)²] / (6 - 1)
= [22.98 + 1.73 + 8.12 + 0.72 + 0.37 + 16.92] / 5
= 50.84 / 5
= 10.17

Step 3: Calculate the F-statistic.

The F-statistic is calculated using the formula:

F = s₁² / s₂²

In this case, F = 20.01 / 10.17 ≈ 1.964

Step 4: Determine the critical value of the F-distribution.

To determine if the variances are significantly different, we need to compare the calculated F-statistic to the critical value of the F-distribution at the given significance level.

Since we are conducting the F-test at the 1% significance level, we need to find the critical F-value with (n₁ - 1) degrees of freedom for Group A and (n₂ - 1) degrees of freedom for Group B.

For Group A, degrees of freedom (df₁) = n₁ - 1 = 6 - 1 = 5
For Group B, degrees of freedom (df₂) = n₂ - 1 = 6 - 1 = 5

Using a statistical table or software, we can find that the critical F-value at the 1% significance level for df₁ = df₂ = 5 is approximately 6.61.

Step 5: Compare the calculated F-statistic with the critical F-value.

If the calculated F-statistic is greater than the critical F-value, we reject the null hypothesis (H0) and conclude that the variances are significantly different. 
Otherwise, if the calculated F-statistic is less than or equal to the critical F-value, we fail to reject the null hypothesis and conclude that there is not enough evidence to suggest a significant difference in variances.

In this case, the calculated F-statistic is 1.964, and the critical F-value at the 1% significance level is 6.61. Since 1.964 < 6.61, we fail to reject the null hypothesis.
Therefore, we conclude that there is not enough evidence to suggest a significant difference in variances between Group A and Group B at the 1% significance level.
```
